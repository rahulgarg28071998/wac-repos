```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
```

```python
X = np.arange(1,5)
plt.plot(X , color= "red" , marker = "*")
plt.plot(X**2)
plt.plot(X**3)
```

```python
plt.scatter(X,X)
plt.scatter(X,X**2)
plt.scatter(X,X**3) 
```

```python
plt.scatter(X , X**2, color= "red" , marker = "*" , label="2018")
plt.scatter(X , X**3, color= "orange" , marker = "*" , label="2020")
plt.xlabel("stock")
plt.ylabel("price")
plt.title("stock price predicition")
# plt.grid(False)
plt.legend()
```
